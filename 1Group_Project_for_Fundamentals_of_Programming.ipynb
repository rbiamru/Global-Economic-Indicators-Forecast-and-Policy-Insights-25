```python
!pip install requests
import requests # Import the requests library at the beginning of your code
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import LocalOutlierFactor
import random

# Assuming your data is in 'economic_data.csv'
df = pd.read_csv('/content/Economic Indicators And Inflation.csv')

# --- Descriptive Statistics ---
def descriptive_statistics(df):
    """Calculates descriptive statistics and visualizes distributions."""
    # Get actual column names from the DataFrame
    numeric_cols = df.select_dtypes(include=np.number).columns.tolist()

    # Measures of central tendency and dispersion
    print("\nDescriptive Statistics:")
    print(df[numeric_cols].describe())
  # Histograms and box plots
    num_cols = len(numeric_cols)  # Get the number of numeric columns
    num_rows = (num_cols + 2) // 3  # Calculate the number of rows needed

    plt.figure(figsize=(15, 5 * num_rows))  # Adjust figure height based on rows
    for i, col in enumerate(numeric_cols):
        plt.subplot(num_rows, 3, i + 1)  # Create subplots dynamically
        sns.histplot(df[col], kde=True)
        plt.title(f"Distribution of {col}")

    plt.tight_layout()  # Adjust subplot spacing
    plt.show()

    plt.figure(figsize=(15, 5 * num_rows))
    for i, col in enumerate(numeric_cols):
        plt.subplot(num_rows, 3, i + 1)
        sns.boxplot(x=df[col])
        plt.title(f"Boxplot of {col}")

    plt.tight_layout()  # Adjust subplot spacing
    plt.show()

# Assuming 'GDP' column contains data with outliers:
data_with_outliers = df[' Unemployment Rate (%)'].values  # Assigning the 'GDP' column values to data_with_outliers

outlier_factor = LocalOutlierFactor()
outliers = outlier_factor.fit_predict(data_with_outliers.reshape(-1, 1))

outliers_mask = outliers == 1 # 1 indicates inliers
outliers_removed = data_with_outliers[outliers_mask]
print(len(outliers_removed))
descriptive_statistics(df)


# --- Trend Analysis ---
import pandas as pd
import matplotlib.pyplot as plt
import random

countries_to_analyze = [
    "USA",
    "China",
    "Japan",
    "Germany",
    "India",
    "UK",
    "Canada",
    "Russia",
    "Australia",
    "France",
    "South Korea",
    "Saudi Arabia",
    "Brazil",
    "Italy",
    "Bangladesh",
    "Indonesia",
    "Turkey",
    "Malaysia",
    "Pakistan",
]
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import random

def trend_analysis(df, country_list=None, country=None):
    """Analyzes trends in GDP, inflation, and unemployment over time for specified countries,
    coloring lines based on average GDP."""
    plt.figure(figsize=(15, 7))

    if country_list is None and country is None:
        country_list = df['Country'].unique()
    elif country is not None:
        country_list = [country]
    elif country_list is not None: # Corrected indentation for this line
        pass

    # Calculate average GDP for each country
    avg_gdp = df.groupby('Country')['GDP'].mean()
    min_avg_gdp = avg_gdp.min()
    max_avg_gdp = avg_gdp.max()

    # Get the 'Blues' colormap
    cmap = cm.get_cmap('Blues')

    for country_name in country_list:
        country_df = df[df['Country'] == country_name]

        # Check if country_df is empty (country not found)
        if country_df.empty:
            print(f"Warning: Country '{country_name}' not found in the data.")
            continue  # Skip to the next country

        # Get the average GDP for the current country
        if country_name in avg_gdp:
            avg_gdp_country = avg_gdp[country_name]
        else:
            print(f"Warning: Average GDP not found for '{country_name}'. Using default color.")
            color_val = 0.5 # Default to a middle shade
        # Normalize the average GDP to the range [0, 1]
        if max_avg_gdp > min_avg_gdp:
            normalized_gdp = (avg_gdp_country - min_avg_gdp) / (max_avg_gdp - min_avg_gdp)
        else:
            normalized_gdp = 0.5 # If all average GDPs are the same

        # Get the color from the 'Blues' colormap
        color = cmap(normalized_gdp)

        #plt.plot(country_df['Year'], country_df['GDP'], label=f'{country_name} GDP', color=color, linestyle='-')
        plt.plot(country_df['Year'], country_df['Inflation'], label=f'{country_name} Inflation', color=color, linestyle='--')
        plt.plot(country_df['Year'], country_df['Unemployment'], label=f'{country_name} Unemployment', color=color, linestyle=':')

    plt.xlabel('Year')
    plt.ylabel('Value')
    title = "Trends in Inflation Rate(%) & Unemployment Rate(%)"
    if country:
        title += f" ({country})"
    elif country_list: # Corrected indentation for this line
        if set(country_list) == set(df['Country'].unique()):
            title +=" (Canada, China & the USA)"
        else:
            title += " (Selected Countries)"
    else:
        title += " (Canada, China & the USA)"
    plt.title(title)

    # Create a legend
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys(), loc='upper left', bbox_to_anchor=(1.05, 1))
    plt.tight_layout() # Adjust layout to prevent labels from overlapping
    plt.show()
# Sample Data (as you provided)
data = {'Country': [], 'Year': [], 'GDP': [], 'Inflation': [], 'Unemployment': []}
countries_to_analyze = ["Canada","China","USA" ]
years = range(2010, 2024) # Reduced years for a smaller example
for country in countries_to_analyze:
    for year in years:
        data['Country'].append(country)
        data['Year'].append(year)
        data['GDP'].append(random.randint(100, 1000))
        data['Inflation'].append(random.uniform(0, 10))
        data['Unemployment'].append(random.uniform(2, 15))
df = pd.DataFrame(data)

# Analyze the specified list of countries with color based on average GDP
trend_analysis(df, country_list=countries_to_analyze)

# You can still analyze a single country (it will have a color based on its average GDP)
# trend_analysis(df, country="USA")

# Or analyze all unique countries (colors will be based on their average GDP)
#trend_analysis(df)



# --- Comparative Analysis ---
def comparative_analysis(df):
    """Compares economic performance across countries using clustering."""
    # Create a copy of the DataFrame to avoid modifying the original
    df_cluster = df.copy()
    # Strip leading/trailing spaces from column names:
    df_cluster.columns = df_cluster.columns.str.strip()
    numerical_data = df_cluster.pivot_table(index='Country', values=['GDP', 'Inflation', 'Unemployment'], aggfunc='mean')
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(numerical_data)

    kmeans = KMeans(n_clusters=3, random_state=42)
    numerical_data['Cluster'] = kmeans.fit_predict(scaled_data)  # Assign clusters to numerical_data

    # Merge cluster labels back into the original DataFrame based on 'Country'
    df_cluster = df_cluster.merge(numerical_data[['Cluster']], on='Country', how='left')

    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='GDP', y='Inflation', hue='Cluster', data=df_cluster)  # Use the updated DataFrame

    plt.title('Country Clustering based on GDP and Inflation')
    plt.show()

# --- Correlation and Regression Analysis ---

def correlation_regression_analysis(df):
    """Analyzes correlations and builds regression models."""
    df.columns = df.columns.str.strip()

    numerical_data = df[['GDP', 'Inflation', 'Unemployment']]
    correlation_matrix = numerical_data.corr()
    print("\nCorrelation Matrix:")
    print(correlation_matrix)

    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap='Blues')
    plt.title('Correlation Heatmap')
    plt.show()

    # Regression example (GDP vs Inflation)
    X = df['Inflation']
    y = df['GDP']
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()
    print("\nGDP vs Inflation Regression Summary:")
    print(model.summary())

correlation_regression_analysis(df)

# --- Time Series Analysis ---
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import random
import statsmodels.api as sm

def time_series_analysis(df, country='China', variable='GDP', start_year=2010, end_year=2024):
    """Performs time series analysis and forecasting for a specific country and variable."""
    country_df = df[df['Country'] == country].sort_values(by='Year')
    country_df = country_df[(country_df['Year'] >= start_year) & (country_df['Year'] <= end_year)]

    if country_df.empty:
        print(f"Warning: No data found for {country} between {start_year} and {end_year} for {variable}.")
        return

    time_series = country_df.set_index('Year')[variable].dropna()

    if len(time_series) < 4:  # ARIMA requires a minimum number of data points
        print(f"Warning: Insufficient data points ({len(time_series)}) for ARIMA analysis for {country} - {variable}.")
        return

    try:
        # ARIMA model
        model = ARIMA(time_series, order=(5, 1, 0))
        model_fit = model.fit()
        print(f"\n--- ARIMA Summary for {country} - {variable} ---")
        print(model_fit.summary())

        # Forecast
        forecast_steps = 5
        forecast = model_fit.forecast(steps=forecast_steps)
        print(f"\n{variable} Forecast for {country} (Next {forecast_steps} Years):")
        print(forecast)

        plt.figure(figsize=(12, 6))
        plt.plot(time_series, label='Observed')

        # Create forecast dates
        last_year = time_series.index[-1]
        forecast_index = range(last_year + 1, last_year + 1 + forecast_steps)
        plt.plot(forecast_index, forecast, label='Forecast', color='#00A189')

        plt.title(f"{variable} Time Series and Forecast for {country} ({start_year}-{end_year})")
        plt.xlabel('Year')
        plt.ylabel(variable)
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Error during ARIMA analysis for {country} - {variable}: {e}")

# Assuming your DataFrame 'df' is already loaded
# Example of how to create a sample DataFrame for testing:
data = {'Country': [], 'Year': [], 'GDP': [], 'Inflation': [], 'Unemployment': []}
years = range(2010, 2025)
countries = ["USA", "China", "Canada"]
for country in countries:
    for year in years:
        data['Country'].append(country)
        data['Year'].append(year)
        data['GDP'].append(random.randint(100, 1000))
        data['Inflation'].append(random.uniform(0, 10))
        data['Unemployment'].append(random.uniform(2, 15))
df = pd.DataFrame(data)

# Analyze GDP for China from 2010-2024
time_series_analysis(df, country='China', variable='GDP', start_year=2010, end_year=2024)
time_series_analysis(df, country='Canada', variable='GDP', start_year=2010, end_year=2024)
time_series_analysis(df, country='USA', variable='GDP', start_year=2010, end_year=2024)


# Analyze Unemployment for China from 2010-2024
time_series_analysis(df, country='China', variable='Unemployment', start_year=2010, end_year=2024)
time_series_analysis(df, country='Canada', variable='Unemployment', start_year=2010, end_year=2024)
time_series_analysis(df, country='USA', variable='Unemployment', start_year=2010, end_year=2024)

# Analyze Inflation for China from 2010-2024
time_series_analysis(df, country='China', variable='Inflation', start_year=2010, end_year=2024)
time_series_analysis(df, country='USA', variable='Inflation', start_year=2010, end_year=2024)
time_series_analysis(df, country='Canada', variable='Inflation', start_year=2010, end_year=2024)


# --- Economic Modeling (Basic Example) ---
def economic_modeling(df):
    """Demonstrates a simple economic model (linear regression)."""
    print(df.columns)  # Print the columns to see their exact names

    # Adjust column names if needed
    X = df[['Inflation', 'Unemployment']]
    y = df['GDP']
    X = sm.add_constant(X)

    model = sm.OLS(y, X).fit()
    print("\nEconomic Model Summary:")
    print(model.summary())

    # Example: Simulating a policy change (e.g., change in inflation)
    new_inflation = df['Inflation'].mean() + 1  # Increase inflation by 1 unit
    new_unemployment = df['Unemployment'].mean()
    predicted_gdp = model.predict([1, new_inflation, new_unemployment])[0]
    print(f"\nPredicted GDP with increased inflation: {predicted_gdp}")

economic_modeling(df)

# Apply economic modeling for all countries in the data
for country in df['Country'].unique():
    country_df = df[df['Country'] == country]
    print(f"\n--- Economic Modeling for {country} ---")
    economic_modeling(country_df)
```
